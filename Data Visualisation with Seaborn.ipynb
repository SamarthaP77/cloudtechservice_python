```python
#Seaborn Plots
#Lm plot

import matplotlib.pyplot as plt
import seaborn as sb 
import pandas as pd

df = sb.load_dataset("anscombe")
print(df)

#Show the results of a linear regression 
sb.Implot(x="x", y="y", data=df)
plt.show()  

```

```python

#Marginal Plot

df sb.load_dataset('iris')
print(df) 
sb.jointplot(x=df["sepal_lengh"], y=df["sepal_width"], kind = 'scatter')
```

```python

#Stripplot
tips = sb.load_dataset('tips')
print(tips.head())
sb.stripplot(x="day", y="total_bill", data=tips)

```

```python
#Swarmplot
tips = sb.load_dataset('tips')
sb.swarmplot(x="day", y="total_bill", data=tips)

```

```python
#Boxplot
sb.boxplot(x="day", y="total_bill", data=tips ,palette='rainbow')
```

```python
#Violinplot
sb.violinplot(x="day", y="total_bill", data=tips ,palette='rainbow')

```
